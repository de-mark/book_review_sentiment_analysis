```python
from splinter import Browser
import pandas as pd
```

```python
browser = Browser("chrome", headless=False)
```

```python
browser.visit("https://www.goodreads.com/choiceawards/best-fiction-books-2022")

all_categories = [cat["href"] for cat in browser.find_by_css(".categoriesList__categoryLink")]
all_booklinks = []

for cat in all_categories:
    browser.visit(cat)
    this_cat_booklinks = [book["href"] for book in browser.find_by_css(".pollAnswer__bookLink")]
    all_booklinks.extend(this_cat_booklinks)

all_booklinks
```

```python
len(all_booklinks)
```

```python
all_reviews = []

for idx, link in enumerate(all_booklinks):
    browser.visit(link)
    print(f"[{idx}/{len(all_booklinks)}] Running {link}", end=" ")
    try:
        this_page_reviews = [review.text for review in browser.find_by_css(".TruncatedContent__text") if review.text != ""]
        all_reviews.extend(this_page_reviews)
    except:
        pass
    print("DONE!")

all_reviews
```

```python
browser.quit()
```

```python
df = pd.DataFrame({"text":all_reviews})
df
```

```python
df.text = df.text.map(lambda t: t.lower().replace("\n", " ").replace("  ", " "))
df.text
```

```python
df.to_csv("../ml_model/data/embed_reviews.csv", index=False)
```
